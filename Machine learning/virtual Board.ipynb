```python
#RUN THIS
import cv2
import mediapipe as mp
import numpy as np

# Initialize the webcam
cap = cv2.VideoCapture(0)
cap.set(3, 1280)  # Set width of the frame
cap.set(4, 720)   # Set height of the frame

# Initialize the hand tracking module
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

# Create a blank canvas
canvas = np.zeros((720, 1280, 3), np.uint8)

# Initialize previous finger position
prev_index_finger_pos = None

# Initialize erase mode flag
erase_mode = False

while True:
    # Read frames from the webcam
    success, frame = cap.read()
    if not success:
        break

    # Convert the frame to RGB for hand tracking
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect hands in the frame
    results = hands.process(frame_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Get the position of index finger tip
            index_finger_pos = (
                int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * frame.shape[1]),
                int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * frame.shape[0])
            )

            # Check if index finger is up (drawing mode)
            if hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y < hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].y:
                # Draw a line on the canvas if previous finger position exists
                if prev_index_finger_pos is not None:
                    if not erase_mode:
                        cv2.line(canvas, index_finger_pos, prev_index_finger_pos, (0, 255, 0), thickness=5)
                    else:
                        cv2.line(canvas, index_finger_pos, prev_index_finger_pos, (0, 0, 0), thickness=20)

            prev_index_finger_pos = index_finger_pos

            # Toggle erase mode if thumb and pinky fingers are up
            thumb_up = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP].y < hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y
            pinky_up = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_TIP].y < hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y
            if thumb_up and pinky_up:
                erase_mode = not erase_mode

    # Resize the canvas to match the frame size
    canvas_resized = cv2.resize(canvas, (frame.shape[1], frame.shape[0]))

    # Display the frame with the canvas
    frame_with_canvas = cv2.add(frame, canvas_resized)
    cv2.imshow("Virtual Whiteboard", frame_with_canvas)
    if cv2.waitKey(1) & 0xFF == ord('c'):
        canvas = np.zeros((720, 1280, 3), np.uint8)
    # Exit loop when 'q' is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
