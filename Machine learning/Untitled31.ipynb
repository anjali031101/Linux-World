```python
import pandas as pd
```

```python
dataset = pd.read_csv("wine.csv")
```

```python
dataset   #3 wines means 3outputs 3 y
#y1 y2 y3   multiclass categorical var  prob percentage
```

```python
y= dataset["Wine"]
```

```python
y
```

```python
y_cat = pd.get_dummies(y)  #multicolinearity issue is not in y no need to drop in this like x1 x2..
```

```python
y_cat
```

```python
dataset.columns
```

```python
X = dataset.drop("Wine",axis=1)
```

```python
X
```

```python
X.shape
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Dense
```

```python
model.get_config()
```

```python
model.add(Dense (
                 kernel_initializer='zeros',
                  units=4,
                  input_shape=(13,),
                  activation='relu',
                  bias_initializer='zeros'
    
))
```

```python
model.get_config()
```

```python
model.summary()
```

```python
model.add(Dense (
                 kernel_initializer='zeros',
                  units=5,
                  activation='relu',
                  bias_initializer='zeros'
    
))
```

```python
model.get_config()
```

```python
model.summary()
```

```python
model.add(Dense (
                 kernel_initializer='zeros',
                  units=3,
                  activation='softmax',
                  bias_initializer='zeros'
    
))
```

```python
model.get_config()
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy")
```

```python
model.fit(X,y_cat,epochs=20)  #AFTER RUNNING FIRST EPOCHS ;FROM SCND BIAS AND WEIGHT WILL RUN FROM THE NEXT VALUE
```
