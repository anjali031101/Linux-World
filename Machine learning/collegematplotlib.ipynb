```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x=np.linspace(1,50,100)
```

```python
y=x**2
z=x**3
```

```python
plt.plot(x,y,marker='o',linestyle ='--',color='g',label='X^2 Data Points')
plt.plot(x,z,marker='o',linestyle ='--',color='r',label='X^3 Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```
